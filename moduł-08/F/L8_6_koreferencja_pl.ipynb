### Przetworzenie tekstu przy użyciu Multiservice (IPI PAN)

```
import requests
import json

url = "http://multiservice.nlp.ipipan.waw.pl/pl"
url_new_request = url + "/addRequest"
url_get_result = url + "/getResult/json/TOKEN"
```

```
text = "To będzie już druga próba licytacji nieruchomości na pl. Słonecznym, którą urzędnicy wytropili po latach poszukiwań majątku Adama Gesslera. "
text += "Jego dług wobec miasta szacują dziś na ok. 27 mln zł. Już w 1992 r., wkrótce po podpisaniu umowy najmu lokalu na Rynku Staromiejskim, zaczęły się problemy z czynszem. "
text += "Sąd orzekł eksmisję. Dotąd miastu udało się odzyskać ledwie kilkadziesiąt tysięcy złotych długu. "
text += "Sprawa budzi wielkie emocje, bo choć Adam Gessler jest słynnym restauratorem, oficjalnie nie ma nic. "
text += "Nawet wynajęta przez Zakład Gospodarowania Nieruchomościami w Śródmieściu firma detektywistyczna nie znalazła majątku. "
text += "Pozostają dwa mieszkania na Żoliborzu, wyceniane przed rokiem na blisko 4,3 mln zł. "
text += "Będą licytowane za dwie trzecie ceny. W ZGN wymyślili, żeby miasto przystąpiło do licytacji. "
text += "Jeśli uda się kupić nieruchomość, komornik pospłaca wierzycieli Adama i Piotra Gesslerów. "
text += "A miasto będzie mogło w przyszłości sprzedać korzystnie atrakcyjny dom. "

request= {"text": text.replace("\n", " "),
           "processingChain":
           [{"serviceName":"Concraft","options":{}},
            {"serviceName":"Spejd","options":{}},
            {"serviceName":"Nerf","options":{}},
            {"serviceName":"MentionStat","options":{}},
            {"serviceName":"Bartek3","options":{}}],
           "clrs":"0"}

r = requests.get(url_new_request, params={"request": json.dumps(request)})

print(r.url)
print(r.text)
```

```
mention_tokens = [(idx, t) for idx, t in enumerate(doc["paragraphs"][0]["sentences"][0]["tokens"]) if t['id'] in mention["childIds"]]
mention_tokens
```

```
span_begin = min([idx for idx, _ in mention_tokens])
span_end = max([idx for idx, _ in mention_tokens])
sentence = ""
for idx, t in enumerate(doc["paragraphs"][0]["sentences"][0]["tokens"]):
  if idx == span_begin:
    sentence += "["
  sentence += t["orth"]
  if idx == span_end:
    sentence += "]"
  sentence += " "
sentence
```

```
from spacy.tokens import Doc, Span
from spacy import displacy

mention_sentence = {}
n = 0
cluster_id = doc["coreferences"][n]["id"]
cluster = doc["coreferences"][n]["mentionIds"]
print(cluster)

for sentence in doc["paragraphs"][0]["sentences"]:
  mention = [m for m in sentence["mentions"] if m["id"] in cluster]

  words = [t["orth"] for t in sentence["tokens"]]
  spaces = [True] * len(sentence["tokens"])
  snt = Doc(nlp.vocab, words=words, spaces=spaces)

  for m in mention:
    segments = m["childIds"]
    indices = []
    for idx, token in enumerate(sentence["tokens"]):
      if token["id"] in segments:
        indices.append(idx)
    span_begin = min(indices)
    span_end = max(indices)
    
    span = Span(snt, span_begin, span_end+1, "mention", cluster_id)
    snt.ents += (span,)


  if len(snt.ents)>0:
    displacy.render(snt, style='ent', jupyter=True, options={'distance': 90})
```
